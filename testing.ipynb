```python
#schema
id_field = FieldSchema(name='id',dtype=DataType.INT64, is_primary=True,auto_id=True)

source_field = FieldSchema(name='source',dtype=DataType.VARCHAR,max_length=255, is_primary=False)

page_field = FieldSchema(name='Page',dtype=DataType.INT64)

TotalPage_field = FieldSchema(name='Total_Page',dtype=DataType.INT64)

embedding_field = FieldSchema(name='embeddings',dtype=DataType.FLOAT_VECTOR,dim=1024)

content_field = FieldSchema(name='content',dtype=DataType.VARCHAR,max_length=2048)
```

```python
schema = CollectionSchema(fields=[id_field,source_field,page_field,TotalPage_field,embedding_field,content_field])
```

```python
Collection=Collection(name='Report',schema=schema)
```

```python
search_params = {
    "metric_type": "L2",
    "params": {"ef": 64}  # ef parameter for HNSW (not efConstruction)
}
```

```python
question = "Introduction to OEE Report for AMTDC Machines"
print(f"Query: '{question}'\n")

# Step 5: Generate query embedding
query_embedding = model.encode([question])

# Step 6: Define search parameters
# search_params = {
#     "metric_type": "L2",      # Use L2 (Euclidean) distance
#     "params": {"nprobe": 10}  # Number of clusters to search
# }

# Step 7: Perform semantic search
results = Collection.search(
    data=query_embedding,
    anns_field="embeddings",  # Your embedding field name
    param=search_params,
    limit=5,  # Top 5 results
    output_fields=["source", "Page", "Total_Page", "content"]  # Return these fields
)

# Step 8: Display results
print("="*70)
print("SEARCH RESULTS")
print("="*70)

for i, hit in enumerate(results[0], 1):
    print(f"\n{i}. [Distance: {hit.distance:.4f}]")
    print(f"   Source: {hit.entity.get('source')}")
    print(f"   Page: {hit.entity.get('Page')} / {hit.entity.get('Total_Page')}")
    print(f"   Content: {hit.entity.get('content')[:200]}...")  # First 200 chars
    print("-" * 70)

print("\n Query complete!")
```
